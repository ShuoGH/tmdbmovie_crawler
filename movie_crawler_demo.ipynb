## Scraping using `bs4`

```python
movie_url="https://www.themoviedb.org/movie/862"
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
blog_first_r = requests.get(movie_url)
```

```python
soup = BeautifulSoup(blog_first_r.text, features="html.parser")
```

```python
soup
```

To get the:
- Movie name 
- poster url
- movie id

which we would use for writing to the image file

```python
# using this method to find the poster url
image_url = soup.find('div', class_='image_content').find('a')['href']
```

```python
image_url
```

```python
# using this code to get the movie title
movie_title=soup.find('div',class_="title").find('h2').text
movie_title
```
